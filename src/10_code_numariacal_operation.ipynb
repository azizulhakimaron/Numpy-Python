```python
import numpy as np
np.random.seed(1000)
arr1=np.array([1,2,3,4])
arr2=np.array([5,6,7,8])
print(arr1+arr2)
print(arr1*arr2)
print(arr1/arr2)
print(arr1**arr2)
matrix1=np.random.randint(1,12,9).reshape(3,3)
matrix2=np.random.randint(1,10,9).reshape(3,3)
print(matrix1)
print(matrix2)
print(matrix1*matrix2)
print(matrix1.dot(matrix2))
```
